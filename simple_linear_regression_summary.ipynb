# Summary

## Simple linear regression

### Model parameters

$$\hat{b}_0 = \overline{y} - \hat{b}_1\overline{x}$$
$$\hat{b}_1 = \frac{S_{xy}}{S_{xx}}$$

$$S_{xx} = \sum_{i=1}^{n}(x_i - \overline{x})^2,$$
$$S_{xy} = \sum_{i=1}^{n} y_i(x_i - \overline{x}).$$



Standard errors:
$$SE(\hat{b}_0) = \sqrt{MS_{Res} \bigg( \frac{1}{n} + \frac{\overline{x}^2}{S_{xx}} \bigg)}$$
$$SE(\hat{b}_1) = \sqrt{\frac{MS_{Res}}{S_{xx}}}$$


### Analysis of variance:

$$SS_T = \sum_{i=1}^{n} (y_i - \overline{y})^2,$$
$$SS_R = \sum_{i=1}^{n} (\hat{y}_i - \overline{y}^2),$$
$$SS_{Res} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2,$$
$$MS_{Res} = \frac{SS_{Res}}{n-2}$$

### Confidence intervals:

**Model parameters:**

$$\hat{b}_0 - t_{a/2, n-2}SE(\hat{b}_0) \leq \hat{b}_0 \leq \hat{b}_0 + t_{a/2, n-2}SE(\hat{b}_0),$$
$$\hat{b}_1 - t_{a/2, n-2}SE(\hat{b}_1) \leq \hat{b}_1 \leq \hat{b}_1 + t_{a/2, n-2}SE(\hat{b}_1).$$


**New prediction for $x=x_0$:**

$$\hat{y}_0 - t_{\alpha/2, n-2}\sqrt{MS_{Res} \bigg( 1+\frac{1}{n} + \frac{(x_0 - \overline{x})^2}{S_{xx}} \bigg)} \leq y_0 \leq \hat{y}_0 + t_{\alpha/2, n-2}\sqrt{MS_{Res} \bigg( 1+\frac{1}{n} + \frac{(x_0 - \overline{x})^2}{S_{xx}} \bigg)}$$


**Mean response for $x=x_0$:**

$$\hat{\mu}_{y|x_0} - t_{\alpha/2, n-2} {\sqrt{MS_{Res} \bigg( \frac{1}{n} + \frac{(x_0 - \overline{x})^2}{S_{xx}}  \bigg)}} \leq E[y|x_0] \leq \hat{\mu}_{y|x_0} + t_{\alpha/2, n-2} {\sqrt{MS_{Res} \bigg( \frac{1}{n} + \frac{(x_0 - \overline{x})^2}{S_{xx}}  \bigg)}}$$

## Multiple linear regression

### Model parameters

The following notation is introduced:

$$y = \begin{bmatrix}
y_1 \\
y_2 \\
... \\
y_n
\end{bmatrix}, \quad 
b = \begin{bmatrix}
b_0 \\
b_1 \\
... \\
b_n
\end{bmatrix}, \quad 
e = \begin{bmatrix}
e_1 \\
e_2 \\
... \\
e_n
\end{bmatrix}, \quad
X = \begin{bmatrix}
1 && x_{11} && x_{12} && ... && x_{1k} \\
1 && x_{21} && x_{22} && ... && x_{2k} \\
... && ... && ... && ... && ... \\
1 && x_{n1} && x_{n2} && ... && x_{nk} \\
\end{bmatrix}.
$$

The least-squares estimator $b$ is:
$$\hat{b} = (X'X)^{-1}X'y,$$

Some helpful matrices:
- $H=X(X'X)^{-1}X'$,
- $C = (X'X)^{-1}$.


### Analysis of variance

For $p=k+1$:

$$SS_T = \sum_{i=1}^{n} (y_i - \overline{y})^2,$$
$$SS_R = \sum_{i=1}^{n} (\hat{y}_i - \overline{y}^2),$$
$$SS_{Res} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2,$$
$$MS_{Res} = \frac{SS_{Res}}{n-p}.$$

Standard errors for $b_j, j=0,1,...,k$ are given as:
$SE(\hat{b}_j) = \sqrt{MS_{Res} \cdot C_{jj}}.$

Adjusted $R^2$:
$$R^2_{adj} = 1 - \frac{SS_{Res}/(n-p)}{SS_T / (n-1)}.$$


## Confidence intervals

**Model parameters:**

$$\hat{b}_j - t_{\alpha/2, n-2} \sqrt{\hat{\sigma}^2C_{jj}} \leq b_j \leq \hat{b}_j + t_{\alpha/2, n-2} \sqrt{\hat{\sigma}^2C_{jj}}, \quad j=0,1,...,k$$

**Mean response**

$$\hat{y}_0 - t_{\alpha/2, n-p}\sqrt{\hat{\sigma}^2x_0'(X'X)^{-1}x_0} \leq E[y | x_0] \leq 
\hat{y}_0 + t_{\alpha/2, n-p}\sqrt{\hat{\sigma}^2x_0'(X'X)^{-1}x_0}$$

**Prediction**

$$\hat{y}_0 - t_{\alpha/2, n-p} \sqrt{\hat{\sigma}^2(1+x_0'(X'X)^{-1}x_0)} \leq y_0 \leq
\hat{y}_0 + t_{\alpha/2, n-p} \sqrt{\hat{\sigma}^2(1+x_0'(X'X)^{-1}x_0)}$$
